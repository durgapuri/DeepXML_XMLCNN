```python
def getData(samples_path,labels_path):
    textfile = open(samples_path, 'r')
    labelfilename = 'label.txt'
    labelfile = open(labels_path,'r')
    vocab = {}
    vc=1
    data = []
    mx = 0
    len_f = 0
    avg = []
    while True: 
        line = textfile.readline() 
        len_f +=1
        line2 = labelfile.readline()
        if not line or not line2: 
            break
        words = line.split()
        words = [word.replace(".", "") for word in words]
        mx = max(mx, len(words))
        #print(words[0])
        avg += [len(words)]
        text = []
        for word in words:
            if word in vocab:
                text.append(word)
            else:
                vocab[word]=vc
                text.append(word)
                vc+=1
        lab  = line2.split()
        #print(lab)
        label = [int(l) for l in lab]
        sample = {}
        # print(text)
        textstring = " , ".join(text)
        sample['text'] = textstring
        sample['catgy'] = label
        data.append(sample)
    print(mx)
    print(len(vocab))
    print(np.mean(avg))
    return data
```

```python
path_eu = '../datasets/EUR-Lex/'
train = getData(path_eu+'train_raw_texts.txt',path_eu+'train_labels.txt')
test = getData(path_eu+'test_raw_texts.txt',path_eu+'test_labels.txt')
```
